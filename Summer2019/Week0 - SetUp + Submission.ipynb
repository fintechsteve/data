## 1.2 - Working with a git branch

Guidelines on modifying your materials for the competition.
1. Open your terminal window and change your current working directory to the folder "imfs-datascience-comp/Summer2019".
2. See all branches in your repo with command **`git branch`**
3. Create a branch with your team name (here, replace XX by your team number). Command **`git branch teamXX_branch`**. If your are team12, the command is **`git branch team12_branch`**
4. Push your branch to the remote repository so we can track it with command **`git push origin teamXX_branch`**
5. Now make sure you are working from your branch with command **`git checkout teamXX_branch`**
6. Every week, we will be releasing new materials. So, make sure to pull latest updates with command **`git pull origin teamXX_branch`**

More info on branching available at https://git-scm.com/book/en/v2/Git-Branching-Basic-Branching-and-Merging  
Also this good link: http://www.gitguys.com/topics/adding-and-removing-remote-branches/  
And this tip https://stackoverflow.com/questions/5601931/what-is-the-best-and-safest-way-to-merge-a-git-branch-into-master

## 1.3 - Submit results

Guidelines on submitting your final results for the competition.
1. You must save your results as a csv following the format per https://github.com/fintechsteve/imfs-datascience-comp/blob/master/Summer2019/templates/securityRecommendation.csv. Your results should be named **"securityRecommendation_teamXX.csv"** where you replace "XX" with your team's number.
2. Open your terminal window and change your current working directory to the folder "imfs-datascience-comp/Summer2019".
3. Change directory to your team's submission folder **`cd submissions/team12`**
4. Import your results csv into the folder per Step3
5. Return to the main folder **`cd IMFS Competition`**
6. Add your changes to your branch **`git add *`**
7. Commit your changes to your branch **`git commit -m "some commit message"`**
8. Push changes locally in your branch **`git push`**
9. Push your branch to the remote repository so we can track it with command **`git push origin teamXX_branch`**
7. Merge your branch to master and push with the following commands:
**`git checkout master`**  
**`git pull origin master`**  
**`git merge teamXX_branch`**  
**`git push origin master`**  
8. Open your web browser and confirm you see your updated results at https://github.com/fintechsteve/imfs-datascience-comp/tree/master/Summer2019/submissions   
9. You're done!! Congratulations!!!
