## 1.1 - Clone competition repository locally

You need to clone the repository to access materials on your local desktop/laptop.
1. Select a directory where you create a folder named "IMFS Competition"
2. Open your terminal window and change your current working directory to the folder you created in Step 1. The command is  **`cd IMFS Competition`**
3. Verify that you have git installed on your computer. In your terminal window, run **`git init`** If error message, download git per instructions at https://product.hubspot.com/blog/git-and-github-tutorial-for-beginners. If you are using a Vanguard Windows desktop/laptop, download git for windows per https://git-scm.com/download/win.
4. Once git is installed, confirm that your current directory is still "IMFS Competition" in your terminal window and run command **`git clone https://github.com/fintechsteve/imfs-datascience-comp/tree/master/Summer2019`**
5. In terminal, run **`cd Summer2019`**
6. In terminal, run **`ls`** You should see materials per image below:

7. you're all set! If error, verify that you have git installed and that you are in the correct directory. Contact IMFS help desk and/or your IMFS mentor for help.
