### Intro
As we start exploring currency returns, we are going to load the data from the repository, visually inspect it and save a copy locally for quick reference later.

Data is stored in a pickle file containing a dictionary, currency_returns, with three arrays:

**currency_returns.data:**  A TxN array of daily currency returns

**currency_returns.dates:** A Tx1 array of datetime arrays indicating the date associated with the rows of the return data

**currency_returns.names:** A 1xN array of currency codes indicating the currency assoicated with the columns of the return data

## Import all necessary libraries for cloud data importing, data manipulation and basic visualization

For this piece, we will need the following packages to be available to our environment:

*   URLLib (for loading data from the web)
*   Cloudpickle (For downloading pickle files from the web)
*   Numpy and Pandas (For data manipulation)
*   DateTime (For basic date manipulation)
*   Matplotlib (For timeseries vizualization)

If the packages are not available, install the with "pip install X"

```python
import urllib.request, urllib.parse, urllib.error
import cloudpickle as cp
import numpy as np, pandas as pd
from datetime import datetime, timedelta
import pickle
import matplotlib.pyplot as plt
```

### Read in data using cp.load and urllib.request.urlopen

*   File is located at https://github.com/fintechsteve/data/blob/master/currency_returns.pkl?raw=true
*   Store the results in currency_returns

```python
currency_returns = cp.load(urllib.request.urlopen('https://github.com/fintechsteve/data/blob/master/currency_returns.pkl?raw=true'))
currency_returns
```

### Use pandas.DataFrame to structure the data as a dataframe
*   Use the 'data' dictionary element for the data.
*   Label the columns with 'names'.
*   Index the rows with 'dates'.
*   Store the results in returns and use .head to visualize the first 10 rows.

```python
returns = pd.DataFrame(currency_returns['data'], columns=currency_returns['names'], index=currency_returns['dates'])
returns.head(10)
```

# is there a better way to add axes labels to the plots.

### Use returns.plot to visualize the data as timeseries
* To plot multiple columns in separate plots create a 9x1 array using plt.subplots. Store the figure as fig, axes as axes.
* Set the ylabel on each axis to the elements in returns.names


```python
fig, axes = plt.subplots(9, 1, figsize=(10, 10))
for axct in range(0,9):
  axes[axct].set_ylabel(returns.columns[axct])
returns.plot(subplots=True, ax=axes, sharex=False, sharey=False, legend=False)
```

### Use .cumsum to plot a cumulative return series

```python
fig, axes = plt.subplots(9, 1, figsize=(10, 10))
for axct in range(0,9):
  axes[axct].set_ylabel(returns.columns[axct])
returns.cumsum().plot(subplots=True, ax=axes, sharex=False, sharey=False, legend=False)
```

### Save the returns matrix locally as ./returns.pkl. We will be using it for the remaining lessons.

```python
returns.to_pickle('./returns.pkl')
```
